# Define a class call Fraction this class is used to represent the ratio of two integers create  a init display include an addition methods equals that takes as import another fraction and returns true if two fraction are identical and false if they are not throw A custom exception if someone tries to create fraction when denominator is zero

```python
class fraction:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def display(self):
        self.f=0
        if type(self.a)==str:
            raise TypeError("a should be integer")
        if type(self.b)==str:
            raise TypeError("b should be integer")
        if self.b==0:
            raise ZeroDivisonError("b should not be Zero")
        self.f=self.a/self.b
        return self.f
    def equals(self,other):
        if(self.f==other.f):
            return True
        else:
            return False
o1=fraction("10",0)
o2=fraction(5,"1")
try:
    o1.display()
    o2.display()
    print(o1.equals(o2))
except Exception as e:
    print(e)        
```
