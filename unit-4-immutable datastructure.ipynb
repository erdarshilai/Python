```python
#craete a function findchr() with declaration def findchr(string,char) this findchr will look for the character in a string and return a index of the first accurence of the character or -1 if the character is not the part of string you cannot use .find and.index 
def findchr(S1,ch):
    for i in range(0,len(s1)):
        if(s1[i]==ch):
            return i
    else:
        return -1
    
s1=input("enter the string:")
ch=input("enter the character for you want to find the location:")            
x=findchr(s1,ch)
print(x)
        
```

```python
#create another function call rfindchr() that will find the last accurrence of character in a string this work simillar to findchr() fun but it starts its search from the end of input string
def rfindchr(S1,ch):
    for i in range(len(s1)-1,-1,-2):
        if(s1[i]==ch):
            return i
    else:
        return -1
    
s1=input("enter the string:")
ch=input("enter the character for you want to find the location:")            
x=rfindchr(s1,ch)
print(x)
```

```python
#create a third function called subchar() with following this descreption (string,orichar,newchar) subchar() function is simillar to findchar() accept whenever this original character found it is replaced by new character the modify string is to be returned use of any substring do not use any inbuilt function
def findchr(S1,oldch,newchar):
    for i in range(0,len(s1)):
        if(s1[i]==oldch):
            return s1[0:i]+newchar+s1[i+1:]
    else:
        return -1
    
s1=input("enter the string:")
oldch=input("enter the oldcharacter for you want too be replaced:")            
newchar=input("enter the new char you want to replace:")
x=findchr(s1,oldch,newchar)
print(x)
```
