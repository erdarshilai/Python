```python
#write a programme to implement caesar cypher techniques for encryption take text as an input from a user and 
#also get the key from user then output should be encrypted text
text=input("enter the text:")
key=int(input("enter the key:"))
l=""
for i in text:
    o1=ord("A")
    o2=ord("a")
    of=0
    if i.isupper():
        of=ord(i)+key
        if of>ord("Z"):
            of=of-26
        l=l+chr(of)
    elif i.islower():
        of=ord(i)+key
        if of>ord("z"):
            of=of-26
        l=l+chr(of)
    else:
        l=l+""
print(l)
        
```
