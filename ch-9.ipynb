```python
#compare two complex number using operaator overloading of (<) operater
class complexNumber:
    def __init__(self,real,imaginary):
        self.real=real
        self.imaginary=imaginary
    def __lt__(self,other):
        if self.real<other.real:
            return True
        elif self.real==other.real:
            if self.imaginary<other.imaginary:
                return True
            else:
                return False
        else:
            return False
comp=complexNumber(1,2)
comp1=complexNumber(3,4)
print(comp.__lt__(comp1))
print(comp1<comp)
```

```python
class complexNumber:
    def __init__(self,real,imaginary):
        self.real=real
        self.imaginary=imaginary
    def __add__(self,other):
        x=self.real+other.real
        y=self.imaginary+other.imaginary
        print("addition:",x,"+j",y)
o1=complexNumber(1,2)
o2=complexNumber(3,4)
print(o1+o2)
```

```python
# from abc import ABC,abstractmethod
class polygon(ABC):
    @abstractmethod
    def noofsides(self):
        pass
class triangle(polygon):
    def noofsides(self):
        print("3 sides")
class pentagon(polygon):
    def noofsides(self):
        print("5 sides")
o1=triangle()
o1.noofsides()
o2=pentagon()
o2.noofsides()
```
