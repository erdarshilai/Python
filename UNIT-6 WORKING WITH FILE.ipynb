```python
#read the data from files and count uppercase letters lowercase letters and spaces
with open("String.txt",'r') as f1:
    data=f1.read()
    print(data)
    u=0
    l=0
    s=0
    for i in data:
        if(i.isupper()):
            u+=1
        if(i.lower()):
            l+=1
        if(i.isspace()):
            s+=1
    print("number of uppercase:",u,"\nnumber of lowercase:",l,"\nnumber of spaces:",s)
```

```python
#to find longest word from the given file text.txt
with open("String.txt",'r') as f1:
    data=f1.read()
    print(data)
    s=data.split()
    l=0
    op=" "
    for i in s:
        k=len(i)
        if k>l:
           l=k
           op=i
    print("length of word:",l)
    print("maximum length word is:",op)
        
    
```

```python
#text processing you are tired of seeing lines on your email-wrap because people type the lines too long for your mail reader appplication
#create a python programme to scan a text file for all the lines longewr than 80 character for each of the offending lines find the closest word before 80 characters and break the line their inserting the remaining text to the new file 
with open("new.txt",'r+') as f1:
    l=f1.read()
    word=l.split()
    s=" "
    p=[]
    for i in range(len(word)):
        if i==0:
            s=word[i]
        else:
            if len(s+word[i])>80:
                p.append(s)
                s=word[i]
            else:
                s+=" "+word[i]
    p.append(s)
with open("new.txt",'w') as f1:
    for i in p:
        f1.write(i)
        f1.write("\n")


    
```
